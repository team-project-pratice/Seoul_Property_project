```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
seoul = pd.read_csv("2022사립고등학교.csv")
seoul
```

```python
seoul2 = seoul.copy()
```

```python
seoul.columns = seoul.iloc[0, :]
```

```python
seoul.drop([0, 1, 2], inplace=True)
```

```python
seoul.reset_index(drop=True, inplace=True)
```

```python
seoul.head(3)
```

```python
new_col = ['자치구별(2)','학교수 (개)']
df = seoul[new_col]
df_1 = df.sort_values(by='학교수 (개)',ascending=False)
df_1
```

```python
seoul_sorted = seoul.sort_values(by='학교수 (개)', ascending=False)
seoul_sorted
```

```python
import matplotlib.pyplot as plt
plt.plot(df_1['자치구별(2)'],df_1['학교수 (개)'])
df_1['학교수 (개)'] = df_1['학교수 (개)'].astype(int)
plt.rc('font',family = 'Malgun Gothic')
plt.xticks(rotation=90)
plt.title('서울 각 자치구별 학교 수')
plt.xlabel('자치구별(2)')
plt.ylabel('학교수 (개)')
plt.show()
```
