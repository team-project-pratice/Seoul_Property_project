```python
import pandas as pd
```

```python
mapo_pd = pd.read_csv('아파트(매매)_실거래가_20241101112021.csv')
```

```python
mapo_pd.head(3)
```

```python
new_col = ['시군구','전용면적(㎡)','거래금액(만원)']
data = mapo_pd[new_col]
data
```

```python
# data['시군구'] = data['시군구'].str.replace('서울특별시','',regex=False)
# data
```

```python
data['시군구2'] = data['시군구'].map(lambda v : v.split(' ')[1])
data
```

```python
print(data.info())
data.describe()
```

```python
data['거래금액(만원)2'] = data['거래금액(만원)'].str.replace(',','')
data
data['거래금액(만원)2'] = pd.to_numeric(data['거래금액(만원)2'])
```

```python
data.info()
data.describe()
```

```python
data2 = data.groupby('시군구2').sum()
```

```python
new_col2 = ['전용면적(㎡)','거래금액(만원)2']
data3 = data2[new_col2]
data3['평당가격(만원)'] = data3['거래금액(만원)2']/data3['전용면적(㎡)']
data4 = data3.sort_values(by='평당가격(만원)',ascending=False)
data4
```
