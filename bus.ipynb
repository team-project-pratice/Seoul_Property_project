```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os

files = os.listdir("data-files/bus_22")
all_df = []
for file in files:
    df = pd.read_csv("data-files/bus_22/{0}".format(file))
    all_df.append(df)

merged_all_df = pd.concat(all_df)
```

```python
on_columns = [col for col in df.columns if '승차' in col]
off_columns = [col for col in df.columns if '하차' in col]
```

```python
df['총승차객수'] = df[on_columns].sum(axis=1)
df['총하차객수'] = df[off_columns].sum(axis=1)
```

```python
result = df.groupby('구')[['총승차객수', '총하차객수']].sum().reset_index()
print(result)
```

```python
result_1 = result[result['구']!='기타'].reset_index(drop=True)
result_2 = result_1.sort_values(by='총승차객수',ascending=False)
result_3 = result_1.sort_values(by='총하차객수',ascending=False)
```

```python
# # 한글 패치
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# # barplot
# sns.barplot( x = df['구'], y = df['총승차객수'])
sns.barplot( x = result_2['구'], y = result_2['총승차객수'])
plt.xticks(rotation=90)
plt.title('22년 서울시 버스 총 승차객수')
plt.xlabel('구')
plt.ylabel('총 승차객수')
plt.show()
```

```python
sns.barplot( x = result_3['구'], y = result_3['총하차객수'])
plt.xticks(rotation=90)
plt.title('22년 서울시 버스 총 하차객수')
plt.xlabel('구')
plt.ylabel('총 하차객수')
plt.show()
```
