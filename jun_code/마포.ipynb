```python
import pandas as pd
```

```python
mapo_pd = pd.read_csv('아파트(매매)_실거래가_20241101112021.csv')
```

```python
mapo_pd.head(3)
```

```python
new_col = ['시군구','전용면적(㎡)','거래금액(만원)']
data = mapo_pd[new_col]
data
```

```python
# data['시군구'] = data['시군구'].str.replace('서울특별시','',regex=False)
# data
```

```python
data['시군구2'] = data['시군구'].map(lambda v : v.split(' ')[1])
data
```

```python
print(data.info())
data.describe()
```

```python
data['거래금액(만원)2'] = data['거래금액(만원)'].str.replace(',','')
data
data['거래금액(만원)2'] = pd.to_numeric(data['거래금액(만원)2'])
```

```python
data.info()
data.describe()
```

```python
data['평 수'] = data['전용면적(㎡)']/3.3
data['평당가격(만원)']  =data['거래금액(만원)2']/data['평 수']
data
new_col = ['시군구2','평당가격(만원)']
data_2 = data[new_col]
data_2
```

```python
data_3 = data_2.groupby('시군구2').mean()
data_4 = data_3.sort_values(by='평당가격(만원)',ascending=False)
data_4
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.bar(data_4.index,data_4['평당가격(만원)'])
plt.rcParams['font.family'] = 'Arial'  # Arial과 같은 한글 지원 폰트로 변경
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호가 깨지는 문제 해결u
plt.rc('font', family='Malgun Gothic')
plt.tight_layout()
plt.xticks(rotation=90)
plt.ylabel('평당가격(만원)')
plt.title('아파트 매매가')
plt.show()
```
