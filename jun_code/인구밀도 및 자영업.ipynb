```python
import pandas as pd
```

```python
data = pd.read_csv("인구밀도_20241104091134.csv")
```

```python
data.head(3)
```

```python
data_sort = data.sort_values(by = '인구밀도 (명/㎢)',ascending = False)
grop_data = data_sort.groupby('동별(2)').sum()
grop_data.drop('동별(1)',axis=1,inplace=True)
grop_data.drop('동별(3)',axis=1,inplace=True)
grop_data.reset_index(inplace=True)
grop_data_1 = grop_data.sort_values(by='인구밀도 (명/㎢)',ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
font_path = "C:\Work\team\malgun-gothic"
plt.rc('font', family='Malgun Gothic')
plt.tight_layout()
plt.figure(figsize=(12,6))
plt.bar(grop_data_1['동별(2)'],grop_data_1['인구밀도 (명/㎢)'])
plt.xticks(rotation=90)
plt.rc('font', family='Malgun Gothic')
plt.title('인구밀도')
plt.show()
```
