```python
import pandas as pd
```

```python
data = pd.read_csv('서울시 부동산 실거래가 정보.csv')
data.head(3)
```

```python
group_data =data.groupby('자치구명').sum()
group_data_1 = group_data.drop('법정동명',axis=1)
group_data_2 = group_data_1.drop('지번구분명',axis=1)
```

```python
new_columns = ['물건금액(만원)', '건물면적(㎡)', '토지면적(㎡)']
group_data_3 = group_data_2[new_columns]
group_data_3
```

```python
new_col = ['접수연도','자치구명','물건금액(만원)','건물면적(㎡)','토지면적(㎡)','건축년도']
data_1 = data[new_col]
data_1[data_1['자치구명']=='중랑구']
```

```python
data_3 = data_1.groupby('자치구명').sum()
data_3
```

```python
data_3['평당가격'] = data_3['물건금액(만원)']/data_3['건물면적(㎡)']
data_4 = data_3.sort_values(by='평당가격',ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.rcParams['font.family'] = 'Arial'  # Arial과 같은 한글 지원 폰트로 변경
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호가 깨지는 문제 해결u
plt.rc('font', family='Malgun Gothic')
plt.tight_layout()
plt.xticks(rotation=90)
plt.title('부동산 실거래 정보')
plt.bar(data_4.index,data_4['평당가격'])
```
