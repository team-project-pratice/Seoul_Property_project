```python
import pandas as pd
```

```python
import chardet

with open('서울시 상권분석서비스(소득소비-자치구).csv', 'rb') as f:
    detector = chardet.universaldetector.UniversalDetector()
    for line in f:
        detector.feed(line) 
        if detector.done:
            break
        detector.close()

    print( detector.result['encoding'] )
```

```python
data = pd.read_csv('서울시 상권분석서비스(소득소비-자치구).csv', encoding='euc-kr')
data
```

```python
# data[data['기준_년분기_코드'] ==20221]

year_mask = data['기준_년분기_코드'].map(lambda v : str(v)[:4] == '2022')
year_mask.sum()
year_data = data[year_mask]
year_data
```

```python
#data.drop('행정동_코드_명', axis=1) / 100000
```

```python
year_data[year_data['행정동_코드_명']=='용산구']
```

```python
grouped_df = year_data.groupby('행정동_코드_명').sum().reset_index()
grouped_df
new_col = ['행정동_코드_명','월_평균_소득_금액','지출_총금액']
grouped_df_1 = grouped_df[new_col]
grouped_df_1['월_평균_소득_금액_1'] = grouped_df_1['월_평균_소득_금액']/4
grouped_df_1['지출_총금액_1'] = grouped_df_1['지출_총금액']/4
grouped_df_1
```

```python
income_df =grouped_df_1.sort_values(by='월_평균_소득_금액_1',ascending=False)
income_df
```

```python
income_df =grouped_df_1.sort_values(by='월_평균_소득_금액_1',ascending=False)
income_df
emplo_df = grouped_df_1.sort_values(by='지출_총금액_1',ascending=False)
emplo_df
```
