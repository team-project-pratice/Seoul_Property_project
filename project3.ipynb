```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
seoul = pd.read_csv("서울시 문화시설 현황.csv")
seoul.info()
```

```python
seoul['지역'].unique()
```

```python
seoul['지역'] = seoul['지역'].str.replace(' ', '')
```

```python
seoul[seoul['지역']=='마포구']
```

```python
jungnang_count = seoul[seoul['지역'] == '중랑구'].shape[0]
print(jungnang_count)
```

```python
data = seoul.groupby('지역')['메인 키'].count()
data.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(data)
plt.rc('font',family = 'Malgun Gothic')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```
