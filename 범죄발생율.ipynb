```python
import pandas as pd
```

```python
data = pd.read_csv('5대+범죄+발생현황_20241105152013.csv',encoding='euc-kr')
```

```python
data
```

```python
data_1 = data.drop(data.index[0])
```

```python
data_1
```

```python
occurrence = data_1.sort_values(by='소계(발생)',ascending=False)
arrest = data_1.sort_values(by='소계(검거)',ascending=False)
occurrence
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

plt.figure(figsize=(10, 5))
width = 0.35
xticks = np.arange(len(occurrence['자치구별(2)']))
plt.bar(xticks - width/2, occurrence['소계(발생)'], width, label='범죄 발생 현황')
plt.bar(xticks + width/2, occurrence['소계(검거)'], width, label = '범죄 검거 현황')
plt.rcParams['font.family'] = 'Arial'  # Arial과 같은 한글 지원 폰트로 변경
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호가 깨지는 문제 해결u
plt.rc('font', family='Malgun Gothic')
plt.tight_layout()
plt.xticks(xticks, occurrence['자치구별(2)'], rotation=90)
plt.ylabel('발생건수')
plt.title('범죄발생 현황 및 검거 현황')
plt.legend()
plt.show()
```
